```python
olympic_data = pd.read_csv('athlete_events.csv')

olympic_data
```

```python
# Group by 'Year', 'Season', and 'Team' and aggregate the data
olympic_aggregate_data = olympic_data.groupby(['NOC', 'Year', 'Season']).agg({
    'ID': 'count',                     # Count of IDs
    'Age': 'mean',                     # Average age
    'Height': 'mean',                  # Average height
    'Weight': 'mean',                  # Average weight
    'Medal': 'count',                  # Count of Medals - non-null values - gold, silver or bronze
    'Sex': lambda x: (x == 'M').sum()  # Number of males
}).reset_index()

olympic_aggregate_data_female = olympic_data.groupby(['NOC', 'Year', 'Season']).agg({
    'ID': 'count',                     # Count of IDs
    'Age': 'mean',                     # Average age
    'Height': 'mean',                  # Average height
    'Weight': 'mean',                  # Average weight
    'Medal': 'count',                  # Count of Medals
    'Sex': lambda x: (x == 'F').sum()  # Number of females
}).reset_index()

# Rename the columns for clarity
olympic_aggregate_data.columns = ['Country', 'Year', 'Season', 'Count_Athlete', 'Avg_Age', 'Avg_Height', 'Avg_Weight', 'Count_Medal', 'Count_Male']
olympic_aggregate_data['Count_Female'] = olympic_aggregate_data_female['Sex']

olympic_aggregate_data = olympic_aggregate_data.fillna(olympic_aggregate_data.mean())

olympic_aggregate_data
```

## Fill missing values

```python
df_filled = final_data.copy()
```

```python
def fill_missing_values(group):
    """
    Fill missing values in each group based on the specified strategy.
    """
    # Sort the group by Year and Season
    group = group.sort_values(by=['Year', 'Season'])

    # Iterate over each column to fill missing values
    for column in group.columns:
        if group[column].isna().any():  # Check if the column has missing values
            # Calculate differences between subsequent rows
            diffs = group[column].diff().dropna()

            # Calculate average difference
            avg_diff = diffs.mean() if not diffs.empty else 0

            # Fill missing values by subtracting the average difference from the next row
            group[column] = group[column].fillna(method='bfill') - avg_diff

    return group

# Apply the function to each country group
df_filled = df_filled.groupby('Country').apply(fill_missing_values).reset_index(drop=True)

df_filled
```
