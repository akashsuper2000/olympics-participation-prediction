```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
olympic_data = pd.read_csv('filled_olympic_data.csv')
olympic_data
```

```python
df = olympic_data.copy()
```

```python
df['Urban_Pop_Ratio'] = df['Urban_Pop'] / df['Pop_Total']
df['Pop_15_64_Ratio'] = df['Pop_15_64'] / df['Pop_Total']
df['Pop_0_14_Ratio'] = df['Pop_0_14'] / df['Pop_Total']

df['Foreign_Invest_GDP_Ratio'] = df['Foreign_Invest'] / df['GDP']

df['Pop_Density'] = df['Pop_Total'] / df['Land_Area']

df['Is_Summer'] = (df['Season'] == 'Summer').astype(int)
```

```python
columns_to_drop = ['Country', 'Year', 'Urban_Pop', 'Pop_0_14', 'Pop_15_64', 'Foreign_Invest', 'Season', 'Avg_Height', 'Avg_Weight', 'Avg_Age']

df.drop(columns=columns_to_drop, inplace=True)
df = df.fillna(df.mean())

df
```

```python
columns_to_scale = [
    'Pop_Total',
    'Net_Migration',
    'Mortality_Rate',
    'Fertility_Rate',
    'Pop_Density',
    'Female_Pop',
    'Electricity',
    'Mobile_Users',
    'Land_Area',
    'GDP',
    'GDP_PCAP',
    'Inflation',
    'GINI',
    'Military_Expense',
    'Unemployment',
    'Labor_Force',
    'Labor_Female',
    'Primary_Completion',
    'Literacy_Rate',
    'Political_Stability',
    'Urban_Pop_Ratio',
    'Pop_15_64_Ratio',
    'Foreign_Invest_GDP_Ratio'
]

scaler = StandardScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

df
```

```python
df.to_csv('train.csv', index=False)
```
