```python
X = olympic_data
Y = olympic_data['Count_Athlete'].values
```

```python
N_Folds =5
kf = KFold(n_splits=N_Folds, shuffle=True, random_state=42)
mae_scores = []
```

```python
mae_scores_xgb = []
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'mae',
    'max_depth': 3,
    'learning_rate': 0.1,
    'n_estimators': 500
}

for fold, (train_idx_xgb, valid_idx_xgb) in enumerate(kf.split(X, Y)):
    X_train_xgb, X_valid_xgb = X.iloc[train_idx_xgb], X.iloc[valid_idx_xgb]
    y_train_xgb, y_valid_xgb = Y[train_idx_xgb], Y[valid_idx_xgb]

    # One-hot encode categorical variables
    X_train_xgb = pd.get_dummies(X_train_xgb)
    X_valid_xgb = pd.get_dummies(X_valid_xgb)

    # Get common feature names
    common_features = set(X_train_xgb.columns) & set(X_valid_xgb.columns)

    # Filter datasets based on common features
    X_train_xgb = X_train_xgb[common_features]
    X_valid_xgb = X_valid_xgb[common_features]

    # Enable categorical support in XGBoost and set feature names
    dtrain = xgb.DMatrix(X_train_xgb, label=y_train_xgb, feature_names=list(common_features), enable_categorical=True)
    dvalid = xgb.DMatrix(X_valid_xgb, label=y_valid_xgb, feature_names=list(common_features), enable_categorical=True)

    model_xgb = xgb.train(params, dtrain, num_boost_round=500, evals=[(dvalid, 'eval')],
                          early_stopping_rounds=50, verbose_eval=50)

    print(f"Fold {fold+1} Training Done.")
    y_pred_valid_xgb = model_xgb.predict(dvalid)
    mae_xgb = mean_absolute_error(y_valid_xgb, y_pred_valid_xgb)
    mae_scores_xgb.append(mae_xgb)

average_mae_xgb = np.mean(mae_scores_xgb)
print(f"5 fold MAE: {average_mae_xgb}")
```

```python
# checking unique categories in categorical features for binary and onehot encoding

unique_categories = {col: len(olympic_data[col].unique()) for col in olympic_data.select_dtypes('object').columns}
unique_categories
```

```python
def onehot_encoder(data, cols):
    df = data.copy()
    for col in cols:
        dummies = pd.get_dummies(df[col],drop_first=True)
        # concatenating dummies and original dataframe
        df = pd.concat([df, dummies], axis=1)

        # dropping original coolumns for which encoding is applied.
        df.drop(col, axis=1,inplace=True)
    return df
```

```python
# applying onehot encoding on features with more than 2 categories
data = onehot_encoder(
    olympic_data,
    cols = ['Country', 'Season']
)
```

```python
# Binary encoding function for categorical features
def binary_encoder(data, columns, positive_values):
    df = data.copy()
    for col, positive_value in zip(columns, positive_values):
        df[col] = df[col].apply(lambda x: 1 if x == positive_value else 0)
    return df
```

```python
# applying Binary encoding on features with 2 categories
data = binary_encoder(
    olympic_data,
    columns=['Season'],
    positive_values=['Summer']
)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Load your dataset
df = olympic_data.copy()

# Separate features and target variable
X = df.drop('Count_Athlete', axis=1)
y = df['Count_Athlete']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2022)

# Define categorical and numerical features
categorical_features = ['Country', 'Season']  # Add other categorical columns as needed
numerical_features = X.columns.difference(categorical_features)

# Create transformers for numerical and categorical features
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Create a preprocessor using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create a pipeline with preprocessing and RandomForestClassifier
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())
])

# Train the model
pipeline.fit(X_train, y_train)

# Make predictions on the test set
y_pred = pipeline.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

```
